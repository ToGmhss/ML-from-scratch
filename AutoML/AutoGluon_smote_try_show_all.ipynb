```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# train_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/train.csv')
# subsample_size = 500  # subsample subset of data for faster demo, try setting this to much larger values
# train_data = train_data.sample(n=subsample_size, random_state=0)
train_data = pd.read_csv("./training_zscore_over.csv")
train_data.head()
```

```python
label = 'class'
print("Summary of class variable: \n", train_data[label].describe())
```

```python
hyperparameters = {
               "NN_TORCH": {},
               'GBM': {},
                #'CAT': {},
               #'CAT': {'iterations': 10000},# cause error!!!
               'RF': {},
               'XT': {},
               'KNN': {},
               #'custom': ['GBM'],
}
```

## Baseline Version

```python
save_path = 'project-1'  # specifies folder to store trained models
eval_metric = 'f1'
predictor = TabularPredictor(label=label, 
                             eval_metric=eval_metric, 
                             path=save_path).fit(train_data,
                                                hyperparameters=hyperparameters)
```

```python
test_data = pd.read_csv("./testing_zscored.csv")#TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/test.csv')
y_test = test_data[label]  # values to predict
test_data_nolab = test_data.drop(columns=[label])  # delete label column to prove we're not cheating
test_data_nolab.head()
```

```python
#predictor = TabularPredictor.load(save_path)  # unnecessary, just demonstrates how to load previously-trained predictor from file

y_pred = predictor.predict(test_data_nolab)
print("Predictions:  \n", y_pred)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
predictor.leaderboard(test_data, silent=True)
```

## Try Advanced Version

### F1 score as the training metric

```python
time_limit = 600  # for quick demonstration only, you should set this to longest time you are willing to wait (in seconds)
metric = 'f1'  # specify your evaluation metric here
predictor = TabularPredictor(label, eval_metric=metric).fit(train_data, 
                                                            time_limit=time_limit, 
                                                            presets='best_quality',
                                                           hyperparameters=hyperparameters)
predictor.leaderboard(test_data, silent=True)
```

```python
y_pred = predictor.predict(train_data, model="LightGBM_BAG_L1")
print("Predictions:  \n", y_pred)
y_train = train_data[label]
perf = predictor.evaluate_predictions(y_true=y_train, y_pred=y_pred, auxiliary_metrics=True)
```

```python
y_pred = predictor.predict(test_data_nolab, model="LightGBM_BAG_L1")
print("Predictions:  \n", y_pred)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
predictor.feature_importance(test_data)
```

### AUC as the training metric

```python
time_limit = 600  # for quick demonstration only, you should set this to longest time you are willing to wait (in seconds)
metric = 'roc_auc'  # specify your evaluation metric here
predictor = TabularPredictor(label, eval_metric=metric).fit(train_data, time_limit=time_limit, presets='best_quality',
                                                           hyperparameters=hyperparameters)
predictor.leaderboard(test_data, silent=True)
predictor.feature_importance(test_data)
```

```python
#y_pred = predictor.predict(test_data_nolab, model="WeightedEnsemble_L2")
pred_probs = predictor.predict_proba(train_data, model="LightGBM_BAG_L1")
#print("Predictions:  \n", y_pred)
print("Predictions:  \n", pred_probs)
#perf = predictor.evaluate_predictions(y_pred=pred_probs)#,#y_test, y_pred=y_pred, auxiliary_metrics=True)
perf = predictor.evaluate_predictions(y_true=y_train,
                                      y_pred=pred_probs, 
                                      auxiliary_metrics=True)
```

```python
#y_pred = predictor.predict(test_data_nolab, model="WeightedEnsemble_L2")
pred_probs = predictor.predict_proba(test_data_nolab, model="LightGBM_BAG_L1")
#print("Predictions:  \n", y_pred)
print("Predictions:  \n", pred_probs)
#perf = predictor.evaluate_predictions(y_pred=pred_probs)#,#y_test, y_pred=y_pred, auxiliary_metrics=True)
perf = predictor.evaluate_predictions(y_true=y_test,
                                      y_pred=pred_probs, 
                                      auxiliary_metrics=True)
```

```python
#y_pred = predictor.predict(test_data_nolab, model="WeightedEnsemble_L2")
pred_probs = predictor.predict_proba(test_data_nolab, model="WeightedEnsemble_L2")
#print("Predictions:  \n", y_pred)
print("Predictions:  \n", pred_probs)
#perf = predictor.evaluate_predictions(y_pred=pred_probs)#,#y_test, y_pred=y_pred, auxiliary_metrics=True)
perf = predictor.evaluate_predictions(y_true=y_test,
                                      y_pred=pred_probs, 
                                      auxiliary_metrics=True)
```
